```python
from transformers import pipeline

# Zero-shot classifier that supports many languages
classifier = pipeline("zero-shot-classification", model="joeddav/xlm-roberta-large-xnli")

def multilingual_ml_prediction(news_text):
    result = classifier(
        news_text,
        candidate_labels=["real", "fake"],
        hypothesis_template="This news is {}."
    )
    label = result["labels"][0]
    print("🔍 ML Prediction:", "✅ Real" if label == "real" else "❌ Fake")
    return label
```
